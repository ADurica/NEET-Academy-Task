```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles = []
prices = []
mileages = []
transmissions = []
months_years = []
fuel_types = []

for page_num in range(1, 120):
    url = 'https://www.autoscout24.hr/lst/volkswagen?atype=C&desc=0&sort=standard&ustate=N%2CU'
    result = requests.get(url)
    content = result.text

    soup = BeautifulSoup(content, 'html.parser')
    listings = soup.find_all('div', {'class': 'ListItem_wrapper__J_a_C'})
    for listing in listings:
        title = listing.find('a', {'class': 'ListItem_title__znV2I ListItem_title_new_design__lYiAv Link_link__pjU1l'}).text.strip()
        price = listing.find('div', {'class': 'PriceAndSeals_wrapper__UStkh'}).text.strip()
        details = listing.find_all('span', {'class': 'VehicleDetailTableNewDesign_item__5LQHk'})
        mileage = details[0].text.strip()
        transmission = details[1].text.strip()
        month_year = details[2].text.strip()
        fuel_type = details[3].text.strip()
        titles.append(title)
        prices.append(price)
        mileages.append(mileage)
        transmissions.append(transmission)
        months_years.append(month_year)
        fuel_types.append(fuel_type)

data = {
    'Title': titles,
    'Price': prices,
    'Mileage': mileages,
    'Transmission': transmissions,
    'Month/Year': months_years,
    'Fuel Type': fuel_types
}
df = pd.DataFrame(data)
display(df)
df.to_csv('autoscout24_data.csv', index=False)
```
